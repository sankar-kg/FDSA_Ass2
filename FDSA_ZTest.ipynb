```python
import pandas as pd
import numpy as np
from statsmodels.stats import weightstats as stests
import math
```

```python
df = pd.read_csv("/content/heart.csv")
DF = df.head()
DF
```

```python
DF.describe()
```

```python
ztest,pvalue = stests.ztest(DF["restecg"],DF["oldpeak"],value=125)
print(float(pvalue))
print(ztest)
if pvalue < 0.05:
  print("Null Rejected and Alternative Accepted")
else:
  print("Null Accepted and Alternative Rejected")
```

```python
ztest,pvalue = stests.ztest(DF["restecg"],value=125)
print(pvalue)
print(ztest)

if pvalue < 0.05:
  print("Null Rejected and Alternative Accepted")
else:
  print("Null Accepted and Alternative Rejected")
```
